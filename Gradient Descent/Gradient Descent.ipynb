# Gradient Descent
***
Gradient descent is a way of determining the lowest value iteratively in a function. When performed over a loop the gradient descent function will constantly descend to a minimum value. This is demonstrated over this specific C function and visualized.

```julia
using Plots
theme(:dark)
```

```julia
C(x) = x^4 - 3x^3 -x^2 + x
```

```julia
plot(C, 
    xlims = (-4, 4),
    xaxis = "x",
    ylims = (-20, 20),
    yaxis = "Cost C(x)",
    legend = false)
```

```julia
ini_w0 = rand((-4:-1))

println("intial weight is $ini_w0")

scatter!([(ini_w0,C(ini_w0))],
        color="blue")
```

```julia
∇C(w) = 1.5*w - 1.5

α = 0.25

w_1 = ini_w0 - α*∇C(ini_w0)

println("ini_w0 = $ini_w0 and |∇C(ini_w0)| = ", abs(∇C(ini_w0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")

plot!([(ini_w0, C(ini_w0)), (w_1, C(w_1))],
        color = "lightblue")
scatter!([(ini_w0, C(ini_w0)), (w_1, C(w_1))],
        color = "yellow")
```

```julia
α = 0.5

w_2 = w_1 - α*∇C(w_1)

println("ini_w0 = $ini_w0 and |∇C(ini_w0)| = ", abs(∇C(ini_w0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")

plot!([(ini_w0, C(ini_w0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "lightblue")
scatter!([(ini_w0, C(ini_w0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "yellow")
```

```julia
α = 0.34

w_3 = w_2 - α*∇C(w_2)

println("ini_w0 = $ini_w0 and |∇C(ini_w0)| = ", abs(∇C(ini_w0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")
println("w_3 = $w_3 and |∇C(w_3)| = ", abs(∇C(w_3)))
println("")

plot!([(ini_w0, C(ini_w0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "lightblue")
scatter!([(ini_w0, C(ini_w0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "yellow")
```

I developed this to demonstrate the difference in gradient descent constantly looking for the min. It has sought the dip on the left side and will decrease no more. Now i will select a different intial weight to demonstrate the gradient will descend towards **A** minimum not the global minimum.
