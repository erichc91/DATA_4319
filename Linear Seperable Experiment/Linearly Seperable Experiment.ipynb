# Linearly Seperable Experiment
****************************************************

```julia
using Plots
```

```julia
coeff = randn(2)
y(x) = coeff[1]*x + coeff[2]

X = randn(100,2)
tuples = [(X[i,1], X[i,2]) for i = 1:100]
tuples[1][1]
```

```julia
above = []
below = []

for point in tuples
    if y(point[1]) < point[2]
        push!(above,point)
    else
        push!(below,point)
    end
end
above[1:5]
```

```julia
scatter!([x for x in above], label = "Upper")
scatter!([x for x in below], label = "Below")
plot!(y, label = "Divider")
```

```julia
function more_points(n)
    coeff = randn(2)
    X = randn(n,2)
    above = []
    below = []
    
    for i = 1:n
        if coeff[1]*X[i,1] + coeff[2] < X[i,2]
            push!(above, (X[i,1], X[i,2]))
        end
        if coeff[1]*X[i,1] + coeff[2] > X[i,2]
            push!(below, (X[i,1], X[i,2]))
        end
    end
    
    return above, below, coeff
end
```

```julia
D1, D2, coeff = more_points(100)
scatter([x for x in D1], label = "Upper")
scatter!([x for x in D2], label = "Lower")
plot!(x->coeff[1]*x + coeff[2], label = "Dividor")
```
